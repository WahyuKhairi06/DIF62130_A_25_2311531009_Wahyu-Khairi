# **bold text**

```python
import pandas as pd

data = {'luas' : [6, 8, 10, 14, 18],
        'harga' : [7, 9, 13, 17.5, 18]}

df = pd.DataFrame(data)
df
```

# **visualisasi data**

```python
import matplotlib.pyplot as plt


df.plot(kind='scatter', x='luas', y='harga')
plt.title('perbandingan Luas Tanah dengan Harga Tanah')
plt.xlabel('Luas Tanah (m persegi)')
plt.ylabel('Harga Tanah (dolar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
import numpy as np

X=np.array(df['luas'])  #independent variable (Featured)
y=np.array(df['harga']) #Dependedn Variabel (target)

print(f'X : {X}')
print(f'y : {y}')
```

```python
X = X.reshape (-1, 1)

print (f'X shape : {X.shape}')
print (f'X: {X}')
```

Modeling

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
x_vis = np.array([0, 25]).reshape(-1, 1)
y_vis = model.predict(x_vis)
```

```python
plt.scatter(X, y)
plt.plot(x_vis, y_vis, color='red')

plt.title('perbandingan Luas Tanah dengan harga Tanah')
plt.xlabel('Luas Tamah (m persegi)')
plt.ylabel('Harga Tanah (dollar)')
#plt.xlim(0, 25)
#plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
print (f'intercept: {model.intercept_} ')
print (f'slope: {model.coef_}')
```

**Prediksi Harga Tanah**

```python
luas_tanah = np.array([12,15,20]).reshape(-1,1)
print(luas_tanah.shape)
luas_tanah
```

```python
harga_prediksi = model.predict(luas_tanah)
harga_prediksi
```

```python
for luas, hrg in zip (luas_tanah, harga_prediksi):
  print (f'Luas Tanah {luas}: harga Prediksi: {hrg}')
```

# **Evaluasi Simple Linear Regression Model**

```python
x_train = np.array([6, 8, 10, 14, 18]).reshape(-1,1)
y_train = np.array([7, 9, 13, 17.5, 18])

x_test = np.array([9, 9, 11, 16, 12]).reshape(-1,1)
y_test = np.array([11, 8.5, 15, 18, 11])
```

Training Simple Linear Regression

```python
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(x_test)
r_squared = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'R-squared: {r_squared}')

```
