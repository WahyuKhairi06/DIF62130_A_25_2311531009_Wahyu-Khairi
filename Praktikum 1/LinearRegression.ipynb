# **1. Import All Library Required**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.datasets import fetch_california_housing
```

Kode diatas adalah implementasi regresi linear dan regresi polinomial untuk memprediksi harga rumah menggunakan dataset California Housing dari Scikit-learn. Berikut adala library yang digunakan

- pandas dan numpy: Digunakan untuk manipulasi data.
- matplotlib.pyplot: Untuk visualisasi data.
- MinMaxScaler: Melakukan normalisasi fitur ke dalam rentang tertentu.
- PolynomialFeatures: Membuat fitur polinomial untuk regresi polinomial.
- train_test_split: Membagi data menjadi set pelatihan dan set pengujian.
- LinearRegression: Model regresi linear.

# **2. Load Dataset**

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['medHouseValue'] = pd.Series(housing.target)

targets=df['medHouseValue'] #dependent variable
inputs=df.drop('medHouseValue',axis=1) #ndependent Variable
df.head(10) #menampilkan 10 data teratas
```

Berikut adalah penjelasan mengenai kode diatas:

1. Mengambil Dataset California Housing

2. Membuat DataFrame Pandas

- pd.DataFrame(data=housing.data, columns=housing.feature_names): Membuat DataFrame Pandas dari data fitur dengan nama kolom yang sesuai.

- df['MedHouseValue'] = pd.Series(housing.target): Menambahkan kolom target (MedHouseValue), yang merupakan harga median rumah dalam ratusan ribu dolar.

3. Memisahkan Variabel Independen & Dependen

- target = df['MedHouseValue']: Menyimpan harga rumah sebagai variabel target (dependent variable).

- inputs = df.drop('MedHouseValue', axis=1): Menghapus kolom MedHouseValue untuk mendapatkan hanya fitur (independent variables).

# **Data Prepocessing**

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.3, random_state=1)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

Kode diatas berfungsi untuk:

1. Membagi Data Menjadi Training dan Testing

train_test_split(inputs, target, test_size=0.3, random_state=1):

- inputs: Variabel independen (fitur) yang digunakan sebagai input.
- target: Variabel dependen (harga rumah).
- test_size=0.3: 30% data digunakan untuk pengujian, 70% untuk pelatihan.
- random_state=1: Menentukan nilai acak agar hasilnya reproducible (tetap sama setiap kali dijalankan).
2. Menampilkan Dimensi Data Setelah Pembagian

- x_train.shape: Menampilkan jumlah baris dan kolom dalam data pelatihan fitur.
- x_test.shape: Menampilkan jumlah baris dan kolom dalam data pengujian fitur.
- y_train.shape: Menampilkan jumlah sampel dalam data pelatihan target.
- y_test.shape: Menampilkan jumlah sampel dalam data pengujian target.

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_test)
```

Kode diatas digunakan untuk:

1. Inisialisasi MinMaxScaler

MinMaxScaler() dari sklearn.preprocessing digunakan untuk menskalakan semua fitur dalam rentang [0,1].

Tujuan: Menyamakan skala fitur agar model bekerja lebih baik dan tidak bias terhadap fitur dengan nilai besar.

2. Menerapkan Scaling ke Data Pelatihan

fit_transform(x_train):

- fit: Menghitung nilai minimum dan maksimum dari data pelatihan.
- transform: Mengubah semua nilai fitur ke dalam rentang [0,1] menggunakan rumus:
Hanya x_train yang di-fit agar skala yang dihitung hanya berdasarkan data pelatihan, menghindari kebocoran data.

3. Menerapkan Scaling ke Data Pengujian

transform(x_test): Menggunakan skala (nilai X_min dan X_max) yang dihitung dari x_train, lalu menerapkan transformasi yang sama ke x_test.

Penting: Jangan gunakan fit_transform(x_test), karena kita tidak ingin menghitung ulang skala berdasarkan data pengujian.

4. Menampilkan Hasil Normalisasi

print(x_test)

# **Modelling**

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

Kode diatas ini melakukan pelatihan model regresi linear menggunakan dataset yang telah dinormalisasi. Berikut penjelasan tiap langkahnya:

1. Inisialisasi Model Regresi Linear

LinearRegression() dari sklearn.linear_model digunakan untuk membuat model regresi linear.

Model ini memprediksi variabel target (y_train) berdasarkan hubungan linear dengan variabel fitur (x_train).

2. Melatih Model dengan Data Pelatihan

fit(x_train, y_train):

- Model mempelajari hubungan linear antara fitur (x_train) dan target (y_train).
- Model akan menentukan koefisien (slope) dan intercept (bias) dari persamaan linear: Screenshot 2025-03-06 104345.png
- Tujuan: Mencari nilai w (koefisien) yang meminimalkan selisih antara prediksi dan nilai sebenarnya (error).

```python
print (f'intercept : {lr_model.intercept_}')
print (f'slope : {lr_model.coef_}')
```

Kode diatas ini digunakan untuk menampilkan parameter dari model regresi linear setelah dilatih. Berikut penjelasan tiap barisnya:

1. Menampilkan Intercept (Bias)

lr_model.intercept_ adalah nilai bias (b) dalam persamaan regresi linear:

Screenshot 2025-03-06 104345.png

Intercept (b) adalah nilai prediksi saat semua fitur bernilai nol.

2. Menampilkan Koefisien (Slope)

lr_model.coef_ berisi array dari bobot (koefisien) untuk setiap fitur.

Setiap nilai dalam array menunjukkan pengaruh perubahan fitur terhadap target.

3. Kesimpulan

intercept_  Nilai prediksi saat semua fitur nol.
coef_  Pengaruh setiap fitur terhadap prediksi harga rumah.
Dapat digunakan untuk menganalisis fitur mana yang paling memengaruhi harga rumah.

# **Model**

```python
y_pred = lr_model.predict(X_test)
mse=mean_squared_error(y_test, y_pred)
r2=lr_model.score(X_test, y_test)
print(f'MSE: {mse}')
print(f'R2: {r2}')
```

Kode diatas ini mengevaluasi performa model regresi linear menggunakan dua metrik utama: Mean Squared Error (MSE) dan R-squared (R). Berikut penjelasannya:

1. Melakukan Prediksi pada Data Uji

predict(x_test): Menggunakan model yang sudah dilatih untuk memprediksi harga rumah (y_pred) berdasarkan fitur dari data uji.

Hasilnya adalah array berisi nilai prediksi untuk setiap sampel di x_test.

2. Menghitung Mean Squared Error (MSE)

mean_squared_error(y_test, y_pred): Menghitung Mean Squared Error (MSE), yang merupakan rata-rata dari kuadrat selisih antara nilai asli (y_test) dan prediksi (y_pred).

MSE yang lebih kecil berarti model lebih akurat.

3. Menghitung R-squared (R)

score(x_test, y_test) menghitung R (koefisien determinasi), yang mengukur seberapa baik model menjelaskan variasi dalam data.

4. Menampilkan Hasil Evaluasi

Menampilkan hasil evaluasi MSE dan R.

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.show()
```

Kode Diatas ini membuat scatter plot untuk membandingkan nilai aktual (y_test) dengan nilai prediksi (y_pred). Berikut penjelasannya:
